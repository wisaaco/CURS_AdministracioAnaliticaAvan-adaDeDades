##  Aprenentatge automàtic (Machine Learning, ML) i Python

Problemes com:<br/>
- detectar quines característiques (agrupament/clustering) tenen en comú un conjunt de persones que han rebut una ajuda,<br/>
- classificar candidats en un concurs de places,<br/>
- etiquetar solars segons el tipus de construcció o ajudes que han rebut,<br/>
- identificar les característiques més rellevants del total per a un determinat segment de la població,<br/>
- donar prioritat automàticament a un conjunt de tasques,<br/>
- reorganitzar intervencions per obtenir el mínim impacte mediambiental dins d'una àrea urbana,<br/>
- etc.<br/>
  
tenen un patró comú que construeix un model emprant dades i utilitzem aquest model per hipòtesi sobre situacions futures. <br/><br/>

La comunitat científica ha seleccionat Python i el seu ecosistema de llibreríes per donar forma a aquest empaquetatge d'algorismes que simplifica el seu ús i el cost computacional associat.<br/>

Els algorismes es classifiquen segons l'ús que se'n fa de les dades. En els algorismes supervisats (*supervised*), hi ha una relació entre l'entrada i la sortida (per exemple, "Medicina gratis" és "SPAM" i "Com et trobes" és "No SPAM"), i s'utilitzen per a la classificació i predicció de noves dades. En els algorismes no supervisats (*unsupervised*), no hi ha informació de sortida associada a les dades, i serveixen per identificar patrons dins les dades. També existeixen algorismes d'aprenentatge per reforç (*reinforcement*), que aprenen funcions mitjançant un procés de simulació amb recompenses. Aquests estan més orientats a problemes d'optimització. D'altra banda, el funcionament de les xarxes neuronals s'utilitza com a algorismes supervisats.<br/>

### Exemple 

Aquest exemple segueix el flux més comú en l'aplicació d'algorismes de ML. Teniu en compte que es requereixen competències bàsiques de Python (i entendre conceptes com *slicing*, ús de funcions i d'altres llibreries com matplotlib o seaborn)!

En el següent codi apliquem un algorisme de regressió (del tipus supervisat):<br/>
- Generant dades fictícies seguint una tendència determinada<br/>
- Afegint valors anormals<br/>
- Definim dos algorismes de regressió: LinearRegression i RANSAC<br/>
- Entrenem cada algorisme amb les dades<br/>
- Dels models obtenim la funció de predicció (una recta amb la tendència que ha après)<br/>
- Avaluació<br/>
- I finalment, una visualització de les dues funcions/models<br/>

Com podem veure del `plot`, l'algorisme RANSAC no es veu afectat pels valors anormals i obté una funció molt alineada amb les dades que aparentment són més coherents amb el problema.


```python
# Referència: https://scikit-learn.org/stable/auto_examples/linear_model/plot_ransac.html#sphx-glr-auto-examples-linear-model-plot-ransac-py (updated 09/2024)

import numpy as np
from matplotlib import pyplot as plt

from sklearn import datasets, linear_model

#######################################################################
# A. Generant dades fictícies seguint una tendència determinada
#######################################################################

n_samples = 1000
n_outliers = 50

X, y, coef = datasets.make_regression(
    n_samples=n_samples,
    n_features=1,
    n_informative=1,
    noise=10,
    coef=True,
    random_state=0,
)

# Add outlier data
np.random.seed(0)
X[:n_outliers] = 3 + 0.5 * np.random.normal(size=(n_outliers, 1))
y[:n_outliers] = -3 + 10 * np.random.normal(size=n_outliers)

#######################################################################
# B. Definim dos algorismes de regressió: LinearRegression i RANSAC
 # I Entrenem cada algorisme amb les dades. FUNCIÓ: fit()
#######################################################################

# Fit line using all data
lr = linear_model.LinearRegression()
lr.fit(X, y)

# Robustly fit linear model with RANSAC algorithm
ransac = linear_model.RANSACRegressor()
ransac.fit(X, y)
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)

#######################################################################
# C. Dels models obtenim la funció de predicció (una recta amb la tendència que ha après)
#######################################################################

# Predict data of estimated models
line_X = np.arange(X.min(), X.max())[:, np.newaxis]
line_y = lr.predict(line_X)
line_y_ransac = ransac.predict(line_X)

#######################################################################
# D. Avaluació i visualització
#######################################################################

# Compare estimated coefficients
print("Estimated coefficients (true, linear regression, RANSAC):")
print(coef, lr.coef_, ransac.estimator_.coef_)

## PLOT
lw = 2
# dades "coherents" /punts verds
plt.scatter(
    X[inlier_mask], y[inlier_mask], color="yellowgreen", marker=".", label="Inliers"
)
# les dadesoutliers / punts grocs
plt.scatter(
    X[outlier_mask], y[outlier_mask], color="gold", marker=".", label="Outliers"
)
# funció de regressió lineal
plt.plot(line_X, line_y, color="navy", linewidth=lw, label="Linear regressor")
# funcio RANSAC
plt.plot(
    line_X,
    line_y_ransac,
    color="red",
    linewidth=lw,
    label="RANSAC regressor",
)
# maquetació del plot
plt.legend(loc="lower right")
plt.xlabel("Input")
plt.ylabel("Response")

plt.show()
```
